##1. Crear una lista que contenga los números impares del 1 al 100

```python
numeros_impares = [num for num in range(1, 101) if num % 2 != 0]
numeros_impares
```

##2. Filtrar los números del 50 al 60

```python
numeros_entre_50_y_60 = [num for num in numeros_impares if 50 <= num <= 60]
numeros_entre_50_y_60
```

##3. Modificar la lista agragando los siguientes 10 números

```python
numeros_impares_add_10 = numeros_impares.copy()
numeros_impares_add_10.extend(range(101,121,2))
numeros_impares_add_10
```

##4. Eliminar los numeros en el los indexados: 16, 24, 32, 11

```python
numeros_impares_after_eliminacion = numeros_impares_add_10.copy()
indices_a_eliminar = [16, 24, 32, 11]
numeros_impares_after_eliminacion = [num for i, num in enumerate(numeros_impares_add_10) if i not in indices_a_eliminar]
numeros_impares_after_eliminacion
```

##5. Organizar de forma descendente la lista

```python
numeros_impares_sort = numeros_impares_after_eliminacion.copy()
numeros_impares_sort.sort(reverse=True)
numeros_impares_sort
```
